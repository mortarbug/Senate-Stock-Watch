```python
import pandas as pd
from pandasql import sqldf
query = lambda q: sqldf(q, globals())
```

```python
 df = pd.read_excel('QuiverQuant Data.xlsx', sheet_name = "Senate Data")
```

I want to see the range of Amounts that senators typically deal with in their stock transactions.

```python
purchases = query("SELECT Stock, Date, Senator, `Purchase / Sale`  as ps FROM df")
```
